```python
# Import dependencies
import pandas as pd
```

```python
# Import rushing Dataset
rushing = pd.read_csv("../Resources/Career_Stats_Rushing.csv")
```

```python
# Visualize dataframe
rushing.head()
```

```python
# Drop columns not needed
rushing.drop(["Player Id", "Name", "Position","Games Played","Rushing Attempts Per Game", "Yards Per Carry", "Rushing Yards Per Game",
              "Longest Rushing Run", "Percentage of Rushing First Downs","Rushing More Than 20 Yards", 
              "Rushing More Than 40 Yards"], axis = 1, inplace = True)
```

```python
# Visualize dataframe
rushing.head()
```

```python
# Get columns
cols = rushing.columns
```

```python
# Value counts
for col in cols:
    print(f"{col}")
    print(rushing[f"{col}"].value_counts())
    print(" ")
```

```python
# Get Stats cols
stats_cols = ["Rushing Attempts", "Rushing Yards", "Rushing TDs", "Rushing First Downs", "Fumbles"]
```

```python
# Replace "--" with "0"
for col in stats_cols:
    rushing[col] = rushing[col].astype(str).str.replace('--','0')
    rushing[col] = rushing[col].astype(str).str.replace(',','').astype(int)
        
```

```python
# Value counts
for col in cols:
    print(f"{col}")
    print(rushing[f"{col}"].value_counts())
    print(" ")
```

```python
# Format year colum
rushing["Year"] = rushing["Year"].astype(str)
```

```python
rushing.head()
```

```python
# Format colnames
rushing.set_axis(["year", "team", "rushing_attempts_rush", "rushing_yards_rush", "rushing_tds_rush",
                 "rushing_first_downs_rush", "fumbles_rush"], axis = 1, inplace = True)
```

```python
# Format team column
rushing["team"] = [row.lower() for row in rushing["team"]]
rushing["team"] = rushing["team"].str.replace('.','')
rushing["team"] = rushing["team"].str.replace(' ','_')
```

```python
# Create unique identifier
rushing["team_year"] = rushing["team"] + "_" + rushing["year"]
```

```python
# Rearrange columns and drop team and year
rushing = rushing[["team_year", "rushing_attempts_rush", "rushing_yards_rush", "rushing_tds_rush", 
                  "rushing_first_downs_rush", "fumbles_rush"]]
```

```python
# Create groupby using unique identifier
rushing_df = rushing.groupby("team_year").sum()
rushing_df
```

```python
# Write clean dataset
rushing_df.to_csv("../Resources/clean_rushing.csv", header = True, index = True)
```
