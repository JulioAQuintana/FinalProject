```python
# Import modules
import pandas as pd
import numpy as np
from functools import reduce
```

```python
# Import one of the datasets
defensive = pd.read_csv("../Resources/Career_Stats_Defensive.csv")
```

```python
# Display head of dataset
defensive.head()
```

```python
# Generate random data for dummy dataset
cols = ["Total Tackles", "Solo Tackles", "Assisted Tackles", "Sacks", "Safties", "Passes Defended", "Ints", "Ints for TDs",
       "Int Yards", "Yards Per Int", "Longest Int Return"]
defensive[cols] = pd.DataFrame(abs(np.random.randn(23998, 11)).round(), columns=list(cols))
```

```python
# Transform dummy data to int type
defensive[cols] = defensive[cols].astype(int)
defensive.dtypes
```

```python
# Drop player id
defensive.drop(["Player Id"], axis=1, inplace = True)
```

```python
# Create unique index column
defensive["index"] = defensive["Team"] + " - " + defensive["Year"].astype(str)
```

```python
# Drop year and groupby unique index column
defensive.drop("Year", axis = 1, inplace = True)
defensive.groupby(["index"]).sum()
```

```python
# Create new year column 
defensive["year"] = defensive["index"].str[-4:].astype(int)
```

```python
# Write to csv
defensive.to_csv("dummy.csv", header = True)
```
