```python
# Import dependencies
import pandas as pd
```

```python
# Import fumble Dataset
fumble = pd.read_csv("../Resources/Career_Stats_Fumbles.csv")
```

```python
# Visualize Dataset
fumble.head()
```

```python
# Drop unecessary columns
fumble.drop(["Player Id", "Name", "Position", "Games Played", "Fumbles", "Own Fumbles Recovered", "Opponent Fumbles Recovered",
            "Out of Bounds Fumbles", "Saftey Fumbles", "Touchbacks"], axis = 1, inplace = True)
```

```python
# Format year colum
fumble["Year"] = fumble["Year"].astype(str)

# Format team column
fumble["Team"] = [row.lower() for row in fumble["Team"]]
fumble["Team"] = fumble["Team"].str.replace('.','')
fumble["Team"] = fumble["Team"].str.replace(' ','_')
```

```python
# Format colnames
fumble.set_axis(["year", "team", "fumbles_lost_fum", "forced_fumbles_fum", "own_fumble_return_yards_fum",
                 "opponent_fumble_return_yards_fum", "fumble_return_tds_fum"], 
                 axis = 1, inplace = True)
```

```python
# Get stats cols
stats_cols = ["fumbles_lost_fum", "forced_fumbles_fum", "own_fumble_return_yards_fum",
                 "opponent_fumble_return_yards_fum", "fumble_return_tds_fum"]
```

```python
# Replace "--" with "0"
for col in stats_cols:
    fumble[col] = fumble[col].astype(str).str.replace('--','0')
    fumble[col] = fumble[col].astype(str).str.replace(',','').astype(int)
```

```python
# Value counts
for col in stats_cols:
    print(f"{col}")
    print(fumble[f"{col}"].value_counts())
    print(" ")
```

```python
# Create unique identifier
fumble["team_year"] = fumble["team"] + "_" + fumble["year"]
```

```python
# Order columns and delete year and team
fumble = fumble[["team_year", "fumbles_lost_fum", "forced_fumbles_fum", "own_fumble_return_yards_fum",
                 "opponent_fumble_return_yards_fum", "fumble_return_tds_fum"]]
```

```python
# Create groupby using unique identifier
fumble_df = fumble.groupby("team_year").sum()
fumble_df
```

```python
# Write clean dataset
fumble_df.to_csv("../Resources/clean_fumbles.csv", header = True, index = True)
```
