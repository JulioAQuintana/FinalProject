```python
# Import dependencies
import pandas as pd
```

```python
# Import receiving Dataset
kickers = pd.read_csv("../Resources/Career_Stats_Field_Goal_Kickers.csv")
```

```python
# Visualize DataFrame
kickers.head()
```

```python
for col in kickers.columns:
    print(col)
```

```python
# Drop unecessary columns
kickers.drop(["Player Id", "Name", "Position", "Games Played", "Longest FG Made", "FG Percentage", 
             "FGs Made 20-29 Yards", "FGs Attempted 20-29 Yards", "FG Percentage 20-29 Yards",
             "FGs Made 30-39 Yards", "FGs Attempted 30-39 Yards", "FG Percentage 30-39 Yards",
             "FGs Made 40-49 Yards", "FGs Attempted 40-49 Yards", "FG Percentage 40-49 Yards",
             "FGs Made 50+ Yards", "FGs Attempted 50+ Yards", "FG Percentage 50+ Yards",
             "Percentage of Extra Points Made"], axis = 1, inplace = True)
```

```python
# Format year colum
kickers["Year"] = kickers["Year"].astype(str)

# Format team column
kickers["Team"] = [row.lower() for row in kickers["Team"]]
kickers["Team"] = kickers["Team"].str.replace('.','')
kickers["Team"] = kickers["Team"].str.replace(' ','_')
```

```python
# Format colnames
kickers.set_axis(["year", "team", "kicks_blocked_kick", "fgs_made_kick", "fgs_attempted_kick",
                 "extra_points_attempted_kick", "extra_points_made_kick","extra_points_blocked_kick"], 
                 axis = 1, inplace = True)
```

```python
# Get Stats cols
stats_cols = ["kicks_blocked_kick","fgs_made_kick", "fgs_attempted_kick",
                 "extra_points_attempted_kick", "extra_points_made_kick","extra_points_blocked_kick"]
```

```python
# Replace "--" with "0"
for col in stats_cols:
    kickers[col] = kickers[col].astype(str).str.replace('--','0')
    kickers[col] = kickers[col].astype(str).str.replace(',','').astype(int)
```

```python
# Value counts
for col in stats_cols:
    print(f"{col}")
    print(kickers[f"{col}"].value_counts())
    print(" ")
```

```python
# Create unique identifier
kickers["team_year"] = kickers["team"] + "_" + kickers["year"]
```

```python
# Create groupby using unique identifier
kickers_df = kickers.groupby("team_year").sum()
kickers_df
```

```python
# Order columns and delete year and team
kickers = kickers[["team_year", "kicks_blocked_kick","fgs_made_kick", "fgs_attempted_kick",
                 "extra_points_attempted_kick", "extra_points_made_kick","extra_points_blocked_kick"]]
```

```python
# Write clean dataset
kickers_df.to_csv("../Resources/clean_field_goal_kickers.csv", header = True, index = True)
```
