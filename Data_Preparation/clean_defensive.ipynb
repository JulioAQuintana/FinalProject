```python
# Import dependencies
import pandas as pd
```

```python
# Import Defensive Dataset
defensive = pd.read_csv("../Resources/Career_Stats_Defensive.csv")
```

```python
defensive.head()
```

```python
# Drop columns not needed; Yards Per Int dropped.
defensive.drop(["Player Id", "Name", "Position", "Yards Per Int", "Longest Int Return"], axis = 1, inplace = True)
```

```python
# Verify dropped columns
defensive.head()
```

```python
# Get Column Names
cols = list(defensive.columns)
```

```python
# Print value counts for columns
for col in cols:
    print(f"{col}")
    print(defensive[f"{col}"].value_counts())
    print(" ")
```

```python
# Get Stats columns
stats_cols = ["Total Tackles", "Solo Tackles", "Assisted Tackles", "Sacks", "Safties", "Passes Defended", 
             "Ints", "Ints for TDs", "Int Yards"]
```

```python
# Replace "--" with "0"
for col in stats_cols:
    defensive[col] = defensive[col].str.replace('--','0')
```

```python
# Print value counts for columns
for col in stats_cols:
    print(f"{col}")
    print(defensive[f"{col}"].value_counts())
    print(" ")
```

```python
# Remove non-sensical decimals points in Sacks column
defensive["Sacks"] = defensive["Sacks"].astype(str).str[0]
```

```python
# Check result from removal of decimal points Sacks
defensive["Sacks"].value_counts()
```

```python
# Reorder Columns
defensive = defensive[["Team", "Year", "Games Played", "Total Tackles", "Solo Tackles", "Assisted Tackles", "Sacks", "Safties", "Passes Defended", 
             "Ints", "Ints for TDs", "Int Yards"]]
defensive.head()
```

```python
# Convert columns to int and year to str
defensive[stats_cols] = defensive[stats_cols].astype(int)
defensive["Year"] = defensive["Year"].astype(str)
defensive.dtypes
```

```python
# Format team column
defensive["Team"] = [row.lower() for row in defensive["Team"]]
defensive["Team"] = defensive["Team"].str.replace('.','')
defensive["Team"] = defensive["Team"].str.replace(' ','_')
```

```python
# Create a unique identifier for team-year
defensive["team_year"] = defensive["Team"] + "_" + defensive["Year"]
defensive.head()
```

```python
# Reorder Columns
defensive = defensive[["team_year", "Team", "Year", "Games Played", "Total Tackles", "Solo Tackles", "Assisted Tackles", "Sacks", "Safties", "Passes Defended", 
             "Ints", "Ints for TDs", "Int Yards"]]
defensive.head()
```

```python
# Rename Columns
defensive.set_axis(["team_year", "team", "year", "games_played", "total_tackles_def", "solo_tackles_def",
                       "assisted_tackles_def","sacks_def", "safties_def", "passes_defended_def", "ints_def", 
                       "ints_for_tds_def", "int_yards_def"], axis = 1, inplace = True)
defensive.head()
```

```python
# Drop games played
defensive.drop(["games_played", "team", "year"], axis = 1, inplace = True)
defensive.head()
```

```python
# Group dataset by unique identifier
defensive_df = defensive.groupby(["team_year"]).sum()
```

```python
# Write clean dataset
defensive_df.to_csv("../Resources/clean_defensive.csv", header = True, index = True)
```
