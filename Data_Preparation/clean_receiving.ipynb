```python
# Import dependencies
import pandas as pd
```

```python
# Import receiving Dataset
receiving = pd.read_csv("../Resources/Career_Stats_Receiving.csv")
```

```python
# Visualize Dataframe
receiving.head()
```

```python
# Drop Columns not needed
receiving.drop(["Player Id", "Name", "Position", "Games Played", "Yards Per Reception", "Yards Per Game",
               "Longest Reception", "Receptions Longer than 20 Yards", "Receptions Longer than 40 Yards"], axis = 1,
              inplace = True)
```

```python
# Get Stats Cols
stats_cols = ["Receptions", "Receiving Yards", "Receiving TDs", "First Down Receptions", "Fumbles"]
```

```python
# Format year colum
receiving["Year"] = receiving["Year"].astype(str)
```

```python
# Value counts
for col in stats_cols:
    print(f"{col}")
    print(receiving[f"{col}"].value_counts())
    print(" ")
```

```python
# Replace "--" with "0"
for col in stats_cols:
    receiving[col] = receiving[col].astype(str).str.replace('--','0')
    receiving[col] = receiving[col].astype(str).str.replace(',','').astype(int)
```

```python
# Value counts
for col in stats_cols:
    print(f"{col}")
    print(receiving[f"{col}"].value_counts())
    print(" ")
```

```python
# Format team column
receiving["Team"] = [row.lower() for row in receiving["Team"]]
receiving["Team"] = receiving["Team"].str.replace('.','')
receiving["Team"] = receiving["Team"].str.replace(' ','_')
```

```python
# Create unique identifier
receiving["team_year"] = receiving["Team"] + "_" + receiving["Year"]
```

```python
# Format colnames
receiving.set_axis(["year", "team", "receptions_rec", "receiving_yards_rec", "receiving_tds_rec",
                 "first_down_receptions_rec", "fumbles_rec","team_year"], 
                 axis = 1, inplace = True)
```

```python
# Rearrange columns and drop team and year
receiving = receiving[["team_year","receptions_rec", "receiving_yards_rec", "receiving_tds_rec",
                 "first_down_receptions_rec", "fumbles_rec"]]
receiving.head()
```

```python
# Create groupby using unique identifier
receiving_df = receiving.groupby("team_year").sum()
receiving_df
```

```python
# Write clean dataset
receiving_df.to_csv("../Resources/clean_receiving.csv", header = True, index = True)
```
