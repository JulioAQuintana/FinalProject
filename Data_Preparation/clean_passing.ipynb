```python
# Import dependencies
import pandas as pd
```

```python
# Import passing Dataset
passing = pd.read_csv("../Resources/Career_Stats_Passing.csv")
```

```python
passing.head()
```

```python
# Drop columns not needed
passing.drop(["Player Id", "Name", "Position", "Games Played", "Completion Percentage", "Pass Attempts Per Game",
             "Passing Yards Per Attempt", "Passing Yards Per Game", "Percentage of TDs per Attempts", 
             "Int Rate", "Longest Pass", "Passes Longer than 20 Yards", "Passes Longer than 40 Yards",
             "Passer Rating"], axis = 1, inplace = True)
passing.head()
```

```python
# Get stats cols
stats_cols = ["Passes Attempted", "Passes Completed", "Passing Yards", "TD Passes", "Ints", "Sacks", "Sacked Yards Lost"]
```

```python
# Value counts
for col in stats_cols:
    print(f"{col}")
    print(passing[f"{col}"].value_counts())
    print(" ")
```

```python
# Replace "--" with "0"
for col in stats_cols:
    passing[col] = passing[col].astype(str).str.replace('--','0')
    passing[col] = passing[col].astype(str).str.replace(',','').astype(int)
```

```python
passing.head()
```

```python
passing.dtypes
```

```python
# Format year colum
passing["Year"] = passing["Year"].astype(str)
```

```python
# Format team column
passing["Team"] = [row.lower() for row in passing["Team"]]
passing["Team"] = passing["Team"].str.replace('.','')
passing["Team"] = passing["Team"].str.replace(' ','_')
```

```python
# Create unique identifier
passing["team_year"] = passing["Team"] + "_" + passing["Year"]
```

```python
# Format colnames
passing.set_axis(["year", "team", "passes_attempted_pass", "passes_completed_pass", "passing_yards_pass",
                 "td_passes_pass", "ints_pass", "sacks_pass", "sacked_yards_lost_pass", "team_year"], 
                 axis = 1, inplace = True)
```

```python
# Rearrange columns and drop team and year
passing = passing[["team_year","passes_attempted_pass", "passes_completed_pass", "passing_yards_pass",
                 "td_passes_pass", "ints_pass", "sacks_pass", "sacked_yards_lost_pass"]]
passing.head()
```

```python
# Create groupby using unique identifier
passing_df = passing.groupby("team_year").sum()
passing_df
```

```python
# Write clean dataset
passing_df.to_csv("../Resources/clean_passing.csv", header = True, index = True)
```
